```python
import numpy as np
from scipy.optimize import curve_fit

# Define a function for nonlinear regression (e.g., logarithmic)
def log_fit(x, a, b):
    return a * np.log(x) + b

# Filter the dataset for the year 2021
data_2021 = df[df['Year'] == 2021]

# Perform nonlinear regression (logarithmic)
x_data = data_2021['AdolFertRate']
y_data = data_2021['LifeExpBirth']
params, _ = curve_fit(log_fit, x_data, y_data)

# Plot the scatter plot with nonlinear regression curve
plt.figure(figsize=(12, 8))
sns.scatterplot(x='AdolFertRate', y='LifeExpBirth', data=data_2021, s=50, color='green')

# Plot the logarithmic regression line
x_line = np.linspace(min(x_data), max(x_data), 100)
y_line = log_fit(x_line, *params)
plt.plot(x_line, y_line, color='red', linewidth=2)

# Add title and labels
plt.title('Nonlinear Regression: Adolescent Fertility Rate vs Life Expectancy in 2021', fontsize=18)
plt.xlabel('Adolescent Fertility Rate (per 1,000 women)', fontsize=14)
plt.ylabel('Life Expectancy at Birth (years)', fontsize=14)
plt.grid(True)
plt.show()
```

```python
import plotly.graph_objects as go

# Filter out rows where Imports%GDP and Exports%GDP are less than 100
filtered_df = df[(df['Imports%GDP'] >= 100) & (df['Exports%GDP'] >= 100)]

# Prepare data for funnel chart from filtered data
stages = ["GDP (Trillions)", "Imports % GDP", "Exports % GDP"]
values = [
    filtered_df['GDP'].mean() / 1e8,  # Convert GDP to trillions
    filtered_df['Imports%GDP'].mean(),  # Keep as percentage
    filtered_df['Exports%GDP'].mean()  # Keep as percentage
]

# Ensure the stages and values are in descending order
sorted_indices = np.argsort(values)[::-1]
stages = np.array(stages)[sorted_indices]
values = np.array(values)[sorted_indices]

# Create the funnel chart
fig = go.Figure(go.Funnel(
    y=stages,
    x=values,
    textinfo="value+percent initial",
    marker=dict(color="lightblue"),
    showlegend=False
))

# Update layout for better visualization
fig.update_layout(
    title='Funnel Chart of Economic Indicators',
    xaxis_title='Value',
    yaxis_title='Stages',
    funnelmode="stack"
)

# Show the plot
fig.show()
```

**12.)Funnel Chart of Economic Indicators**

**Questions Answered:**

1.)How do economic indicators like GDP, imports, and exports flow through different stages?

2.)What is the relationship between these indicators and how do they change?

**Observations:**

1.)The funnel chart shows the progression of economic indicators, from GDP to imports and exports.

2.)Helps visualize how different economic stages compare and where significant changes occur.

3.)Useful for understanding how economic variables impact each other and where major shifts happen.
